```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import ListedColormap
import statsmodels.api as sm
import time
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score, mean_squared_error
from sklearn.model_selection import LeaveOneOut, GridSearchCV, cross_val_score

random_state = 59
```

```python
f = "CKEY_TLS_10x10.xyzch"
df = pd.read_csv(f,
                 header=0)

# localize coordinates
df["X"] = df["X"] - df["X"].min()
df["Y"] = df["Y"] - df["Y"].min()
df["Z"] = df["Z"] - df["Z"].min()

sigma_g = []
sigma_ng = []
H_ng = []

for i in range(30):
    df = df.sample(20000, random_state=random_state)
    
    x_center = df['X'].mean()
    y_center = df['Y'].mean()
    
    df_g = df.loc[df['Classification'] == 2.0]
    n_ground = df_g.count()[0]
    
    df_ng = df.loc[df['Classification'] != 2.0]
    n_nonground = df_ng.count()[0]
    
    sigma_g.append(np.around(df_g.std()[2],3))
    sigma_ng.append(np.around(df_ng.std()[4],3))
    
    Z_ground = df_g['Z'].values
    XY_ground = df_g[['X', 'Y']].values
    XY_ground = sm.add_constant(XY_ground)
    g_model = sm.OLS(Z_ground, XY_ground)
    g_results = g_model.fit()

    Z_nonground = df_ng['Z'].values
    XY_nonground = df_ng[['X', 'Y']].values
    XY_nonground = sm.add_constant(XY_nonground)
    ng_model = sm.OLS(Z_nonground, XY_nonground)
    ng_results = ng_model.fit()
    
    H_ng.append(ng_results.predict(np.array([1, x_center, y_center])) - 
                g_results.predict(np.array([1, x_center, y_center])))

print(f)
print(f"sigma ground = {np.around(np.mean(sigma_g),3)} +/- {np.around(np.std(sigma_g),3)}")
print(f"sigma non-ground = {np.around(np.mean(sigma_ng),3)} +/- {np.around(np.std(sigma_ng),3)}")
print(f"Height of non-ground plane = {np.around(np.mean(H_ng),3)} +/- {np.around(np.std(H_ng),3)}")
print()

# save results for plotting later
df_CKEY = df
gplane_CKEY = g_results
ngplane_CKEY = ng_results
```

```python
# grid search for best model hyperparameters

X_train = np.array([i[0:3] for i in corpus])
n_train = np.array([i[3] for i in corpus])
z0_train = np.array([i[4] for i in corpus])

params_to_test = {
    'n_estimators':[51,101,201,401,801,1601],
    'max_depth':[3,4,5],
    'min_samples_leaf':[1,2,3],
    'max_features':[0.3, 0.6, 0.9, "sqrt", "log2", None]
                 }

rf_model_grid_n = RandomForestRegressor(random_state=random_state, oob_score=True)
n_grid_search = GridSearchCV(rf_model_grid_n, param_grid=params_to_test, refit=True)

n_grid_search.fit(X_train, n_train)
n_best_params = n_grid_search.best_params_
print(n_best_params)

rf_model_grid_z0 = RandomForestRegressor(random_state=random_state, oob_score=True)
z0_grid_search = GridSearchCV(rf_model_grid_z0, param_grid=params_to_test, refit=True)

z0_grid_search.fit(X_train, z0_train)
z0_best_params = z0_grid_search.best_params_
print(z0_best_params) 
```

```python
# ORIGINAL DATA
corpus_O = corpus[:-3]

z0_true_O = []
n_true_O = []
z0_pred_O = []
n_pred_O = []

for s in range(len(corpus_O)):
    X_train = [i[0:3] for i in corpus_O]
    n_train = [i[3] for i in corpus_O]
    z0_train = [i[4] for i in corpus_O]
    
    X_test = np.array(X_train[s]).reshape(1,-1)
    n_true_O.append(n_train[s])
    z0_true_O.append(z0_train[s])
    
    del X_train[s]
    del n_train[s]
    del z0_train[s]
    
    rf_n = RandomForestRegressor(**n_best_params, random_state=random_state, oob_score=True)
    rf_z0 = RandomForestRegressor(**z0_best_params, random_state=random_state, oob_score=True)

    rf_n.fit(X_train, n_train)
    rf_z0.fit(X_train, z0_train)
    
    n_pred_O.append(rf_n.predict(X_test)[0])
    z0_pred_O.append(rf_z0.predict(X_test)[0])
    
n_residuals_O = np.array(n_true_O) - np.array(n_pred_O)
z0_residuals_O = np.array(z0_true_O) - np.array(z0_pred_O)

print("n metrics original data...")
print(f"MAE = {np.around(mean_absolute_error(n_true_O, n_pred_O),5)}")
# print(f"MAPE = {np.around(mean_absolute_percentage_error(n_true_O, n_pred_O),5)}")
print(f"RMSE = {np.around(mean_squared_error(n_true_O, n_pred_O, squared=False),5)}")
print(f"R2 = {np.around(r2_score(n_true_O, n_pred_O),5)}")
print()

print("z0 metrics original data...")
print(f"MAE = {np.around(mean_absolute_error(z0_true_O, z0_pred_O),5)}")
# print(f"MAPE = {np.around(mean_absolute_percentage_error(z0_true_O, z0_pred_O),5)}")
print(f"RMSE = {np.around(mean_squared_error(z0_true_O, z0_pred_O, squared=False),5)}")
print(f"R2 = {np.around(r2_score(z0_true_O, z0_pred_O),5)}")
```

```python
# ORIGINAL DATA + NEW DATA
start = time.time()

z0_true_N = []
n_true_N = []
z0_pred_N = []
n_pred_N = []
n_pred_pert = []
z0_pred_pert = []

n_residuals_stdev = []
z0_residuals_stdev = []

for s in range(len(corpus)):
    X_train = [i[0:3] for i in corpus]
    n_train = [i[3] for i in corpus]
    z0_train = [i[4] for i in corpus]

    X_test = np.array(X_train[s]).reshape(1,-1)
    n_true_N.append(n_train[s])
    z0_true_N.append(z0_train[s])
    
    del X_train[s]
    del n_train[s]
    del z0_train[s]
    
    rf_n = RandomForestRegressor(**n_best_params, random_state=random_state, oob_score=True)
    rf_z0 = RandomForestRegressor(**z0_best_params, random_state=random_state, oob_score=True)
    
    rf_n.fit(X_train, n_train)
    rf_z0.fit(X_train, z0_train)
    
    n_pred_N.append(rf_n.predict(X_test)[0])
    z0_pred_N.append(rf_z0.predict(X_test)[0])
    
    # perturbation test
    n_pred_pert_rep = []
    z0_pred_pert_rep = []
    n_pert_residuals = []
    z0_pert_residuals = []
    for p in range(30):
        rf_n_pert = RandomForestRegressor(**n_best_params, random_state=random_state, oob_score=True)
        rf_z0_pert = RandomForestRegressor(**z0_best_params, random_state=random_state, oob_score=True)
        n_train_pert = np.abs(np.random.normal(n_train, [v/2.0 for v in n_train]))
        z0_train_pert = np.abs(np.random.normal(z0_train, [v/2.0 for v in z0_train]))
        rf_n_pert.fit(X_train, n_train_pert)
        rf_z0_pert.fit(X_train, z0_train_pert)
        n_pred_pert_rep.append(rf_n_pert.predict(X_test)[0])
        z0_pred_pert_rep.append(rf_z0_pert.predict(X_test)[0])
        
    n_pred_pert.append(n_pred_pert_rep)
    z0_pred_pert.append(z0_pred_pert_rep)
    
    n_pert_residuals.append([n_true_N[s] - n_pred_pert[s][i] for i in range(30)])
    z0_pert_residuals.append([z0_true_N[s] - z0_pred_pert[s][i] for i in range(30)])
    
n_residuals_N = np.array(n_true_N) - np.array(n_pred_N)
z0_residuals_N = np.array(z0_true_N) - np.array(z0_pred_N)

n_residuals_stdev = [np.std(r) for r in n_pert_residuals]
z0_residuals_stdev = [np.std(r) for r in z0_pert_residuals]

print("n metrics original+new data...")
print(f"MAE = {np.around(mean_absolute_error(n_true_N, n_pred_N),5)}")
# print(f"MAPE = {np.around(mean_absolute_percentage_error(n_true_N, n_pred_N),1)}")
print(f"RMSE = {np.around(mean_squared_error(n_true_N, n_pred_N, squared=False),5)}")
print(f"R2 = {np.around(r2_score(n_true_N, n_pred_N),5)}")

print("z0 metrics original+new data...")
print(f"MAE = {np.around(mean_absolute_error(z0_true_N, z0_pred_N),5)}")
# print(f"MAPE = {np.around(mean_absolute_percentage_error(z0_true_N, z0_pred_N),1)}")
print(f"RMSE = {np.around(mean_squared_error(z0_true_N, z0_pred_N, squared=False),5)}")
print(f"R2 = {np.around(r2_score(z0_true_N, z0_pred_N),5)}")

end = time.time()
print(np.around((end-start)/60.0))
```
