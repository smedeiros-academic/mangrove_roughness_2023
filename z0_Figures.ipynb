```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
fig, ax = plt.subplots(figsize=(7,5))

f1 = "CKEY_replicates.test"
df1 = pd.read_csv(f1,
                  header=None,
                 names=["PC_fraction", 
                        "r0", "r1", "r2", "r3", "r4", "r5", "r6", "r7", "r8", "r9",
                        "num_pts", "ppsm", "num_vox"])
df1['mean_z0'] = df1.iloc[:, 1:11].mean(axis=1)
df1['errorbar'] = df1.iloc[:, 1:11].std(axis=1)

ax.errorbar(df1.ppsm, df1.mean_z0, yerr=df1.errorbar,
            color="#00549A",
            lw=1.0, elinewidth=1.0, capsize=3,
            label="CKEY")

f2 = "IEPK_replicates.test"
df2 = pd.read_csv(f2,
                  header=None,
                  names=["PC_fraction", 
                        "r0", "r1", "r2", "r3", "r4", "r5", "r6", "r7", "r8", "r9",
                        "num_pts", "ppsm", "num_vox"])
df2['mean_z0'] = df2.iloc[:, 1:11].mean(axis=1)
df2['errorbar'] = df2.iloc[:, 1:11].std(axis=1)

ax.errorbar(df2.ppsm, df2.mean_z0, yerr=df2.errorbar,
            color="darkgoldenrod",
            lw=1.0, elinewidth=1.0, capsize=3,
            label="IEPK")

f3 = "PKEY_replicates.test"
df3 = pd.read_csv(f3,
                  header=None,
                  names=["PC_fraction", 
                        "r0", "r1", "r2", "r3", "r4", "r5", "r6", "r7", "r8", "r9",
                        "num_pts", "ppsm", "num_vox"])
df3['mean_z0'] = df3.iloc[:, 1:11].mean(axis=1)
df3['errorbar'] = df3.iloc[:, 1:11].std(axis=1)

ax.errorbar(df3.ppsm, df3.mean_z0, yerr=df3.errorbar,
            color="black",
            lw=1.0, elinewidth=1.0, capsize=3,
            label="PKEY")

ax.set_xticks([x for x in range(0,300000,10000)])
ax.set_yticks([y for y in np.arange(1.9,2.6,0.1)])
ax.grid(which="both")
ax.set_xscale("linear")
ax.set_xlabel("Point Cloud Density (ppsm)")
ax.set_ylabel("$z_0\ from\ TLS\ (m)$")
ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.legend()

plt.tight_layout()
plt.savefig("TLS_ppsm_sensitivity2.png",dpi=600)
;
```
